```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns

import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
strava = pd.read_csv("strava_activity.csv")
```

```python
strava.head()
```

```python
heart_rate = strava[['athlete.sex', 'average_heartrate', 'average_speed', 'elev_high', 'elev_low', 'max_heartrate', 'athlete.country', 'type']].copy()
```

```python
heart_rate.head(5)
```

```python
heart_rate = heart_rate.dropna()
```

```python
heart_rate.head(5)
```

```python
heart_rate_summary = heart_rate.groupby(['athlete.sex']).sum()
```

```python
heart_rate_summary
```

```python
ride = heart_rate[heart_rate['type'].str.match('Ride')]
```

```python
ride.head(5)
```

```python
male = ride[ride['athlete.sex'].str.match('M')]
male.head()
```

```python
female = ride[ride['athlete.sex'].str.match('F')]
female.head()
```

```python
male_max_heart_rate = male[male['max_heartrate'] > 200]
male_max_heart_rate.sort_values(by=['max_heartrate'], ascending=False, inplace=True)
male_max_heart_rate.head(5)
```

```python
male_max_heart_rate_country = male_max_heart_rate.groupby(['athlete.country']).count()
```

```python
male_max_heart_rate_country.head(5)
```

```python
male_max_heart_rate_country = male_max_heart_rate_country.reset_index()
```

```python
male_max_heart_rate_country.sort_values(by=['athlete.sex'], ascending=False, inplace=True)
```

```python
male_max_heart_rate_country.head(5)
```

```python
female_max_heart_rate = female[female['max_heartrate'] > 200]
female_max_heart_rate.sort_values(by=['max_heartrate'], ascending=False, inplace=True)
female_max_heart_rate.head(5)
```

```python
female_max_heart_rate_country = female_max_heart_rate.groupby(['athlete.country']).count()
female_max_heart_rate_country = female_max_heart_rate_country.reset_index()
female_max_heart_rate_country.sort_values(by=['athlete.sex'], ascending=False, inplace=True)
female_max_heart_rate_country.head(5)
```
