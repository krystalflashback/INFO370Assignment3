```python
#question 1
#do men tend to exercise more intensely than woman? (taking into account both distance and speed)
exercise = strava[['athlete.sex', 'distance', 'average_speed']].copy()
exercise = exercise[~(exercise == 0).any(axis=1)]
exercise.head()
cal_exercise = exercise.groupby(['athlete.sex']).describe()
cal_exercise
#By calculating the mean distance and speed group by sex, we get that the mean distance for female is 16350.02 which is 
#less than the mean distance for male which is 27965. The mean speed for men is also quicker.
#The result shows that men tend to exercise more intensely than woman
```
